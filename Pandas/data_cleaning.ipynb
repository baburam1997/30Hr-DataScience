```python
df = pd.read_csv(csv_file)
df.head()
```

```python
df["datetime"]= pd.to_datetime(df["date_column"])
df["date"] = pd.to_datetime(df["datetime"],format="%y-%m-%d").dt.date
df.head()
```

```python
df["time"] = pd.to_datetime(df["datetime"],format="%d-%m-%y")
df.head()
# https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
```

```python
df["time1"] = pd.to_datetime(df["datetime"],format="%H:%M:%S").dt.strftime('%H:%M:%S')
df.head()
```

```python
df["time2"] = pd.to_datetime(df["datetime"],format="%H:%M:%S").dt.strftime('%I:%M:%S:%p')
df.head()
```

```python
df['year']= df["datetime"].dt.year
df['month']=df["datetime"].dt.month
df['day']=df["datetime"].dt.day
df.head()
```

```python
df['marks']=df['score_1'] + df['score_2']
df.head()
```

```python
df["percentage"]=df["marks"]/200*100
df.head()
```

```python
df['full_name']=df['first_name'] +" "+ df['last_name']
df.head()
```
