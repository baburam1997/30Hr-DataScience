```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
y = [20, 25, 35, 55, 10,  55, 57, 78, 58 ]

plt.plot(x, y)
plt.title("Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Scatter Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
plt.bar(x, y, color='green', edgecolor='blue', 
        linewidth=2)
plt.title("Bar Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
plt.plot(x, y, color='green', linewidth=3, marker='o', markersize=15, linestyle='--')
plt.title("Linear graph")
plt.legend(["Sales"])
plt.show()
```

```python
plt.hist(x, bins=9, color='green', edgecolor='blue', linestyle='--', alpha=0.5)
plt.title("Histogram")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
data = {
    'Date': pd.date_range(start='2024-09-01', periods=7, freq='D'),
    'Day': ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',"Saturday"],
    'Tips': np.random.randint(50, 500, size=7)
}
df_tips = pd.DataFrame(data)
df_tips
```

```python
date = data['Date']
day = data['Day']
tips = data['Tips']
```

```python
plt.plot(date, tips, color='green', linewidth=3, marker='o', markersize=10, linestyle='--')
plt.title("Date vs Tips")
plt.ylabel('Tips')
plt.xlabel('Dates')
plt.legend(["Tips"])
plt.show()
```

```python
plt.plot(day, tips, color='green', linewidth=3, marker='o', markersize=10, linestyle='--')
plt.title("Day vs Tips")
plt.ylabel('Tips')
plt.xlabel('Day')
plt.legend(["Tips"])
plt.show()
```

```python
plt.scatter(date, tips)
plt.title("Scatter Chart for Date vs Tips")
plt.ylabel('Tips')
plt.xlabel('Date')
plt.show()
```

```python
plt.scatter(day, tips)
plt.title("Scatter Chart for Day vs Tips")
plt.ylabel('Tips')
plt.xlabel('Day')
plt.show()
```

```python
plt.bar(date, tips, color='green', edgecolor='blue', linewidth=2)
plt.title("Bar Chart for Date vs Tips")
plt.ylabel('Tips')
plt.xlabel('Date')
plt.show()
```

```python
plt.bar(day, tips, color='green', edgecolor='blue', linewidth=2)
plt.title("Bar Chart for Day vs Tips")
plt.ylabel('Tips')
plt.xlabel('Day')
plt.show()
```

```python
plt.hist(tips, bins=10, color='green', edgecolor='blue', linestyle='--', alpha=0.5)
plt.title("Histogram for tips")
plt.ylabel('Values')
plt.xlabel('Tips')
plt.show()
```
