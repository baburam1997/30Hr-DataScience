Handle Outliers

```python
#Using median calculations and IQR, outliers are identified and these data points should be removed
Q1 = df["M01AB"].quantile(0.25)
Q3 = df["M01AB"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[df["M01AB"].between(lower_bound, upper_bound)]
```

Dataframe splitting for the different purposes


Split by Time (Year/Month/Date Range)

```python
df['datum'] = pd.to_datetime(df['datum'])

# Split by date range
train_df = df[df['datum'] < '2016-01-01']
test_df = df[df['datum'] >= '2016-01-01']
```

Split by Percentage

```python
train_size = int(0.6 * len(df))
val_size = int(0.2 * len(df))

train_df = df[:train_size]
val_df = df[train_size:train_size + val_size]
test_df = df[train_size + val_size:]
```

Split by its size using iloc and loc

```python
df_length = len(df)
# Define split points
train_size = int(0.5 * df_length)
test_size = int(0.8 * df_length)
predict_size = int(0.9 * df_length)
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:test_size]
pred_df = df.iloc[test_size: predict_size]
predict_df = df.iloc[predict_size:]
```
